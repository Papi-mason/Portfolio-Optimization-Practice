```python
#Price and Returns
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get data for IBM from Yahoo!Finance
ibm_data = yf.download("IBM", start="2007-01-01", end="2022-12-31")

# Check if 'Adj Close' column exists, otherwise use 'Close'
# The 'Adj Close' column accounts for splits and dividends, but 'Close' is a good fallback.
if 'Adj Close' in ibm_data.columns:
    ibm_prices = ibm_data['Adj Close']
else:
    ibm_prices = ibm_data['Close']
    print("Warning: 'Adj Close' not found. Using 'Close' prices instead.")

# Calculate log returns
# The diff() function in pandas calculates the difference between consecutive elements
# log() is the natural logarithm
ibm_logreturns = np.log(ibm_prices / ibm_prices.shift(1)).dropna()

# Calculate linear returns (percentage change)
ibm_linreturns = ibm_prices.pct_change().dropna()

# Plot IBM adjusted/close price
plt.figure(figsize=(12, 6)) # Set figure size for better readability
plt.plot(ibm_prices.index, ibm_prices.values, linewidth=0.8, color="darkblue")
plt.title("IBM Stock Price (Adjusted/Close)", fontsize=16)
plt.xlabel(None) # Remove x-axis label as dates are self-explanatory
plt.ylabel(None) # Remove y-axis label
plt.grid(True, linestyle='--', alpha=0.7) # Add a grid for better readability
plt.xticks(rotation=45) # Rotate x-axis labels for better visibility
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

print("\nIBM Log Returns (first 5 rows):")
print(ibm_logreturns.head())

print("\nIBM Linear Returns (first 5 rows):")
print(ibm_linreturns.head())
```

```python
#IBM Rolling Volatility (Left-aligned and Centered)
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Get Data for IBM ---
# Using 'Adj Close' for prices as it accounts for splits and dividends.
# We'll use a longer period to better observe volatility changes.
ibm_data = yf.download("IBM", start="2010-01-01", end="2024-01-01")

# Check if 'Adj Close' column exists, otherwise use 'Close'
if 'Adj Close' in ibm_data.columns:
    ibm_prices = ibm_data['Adj Close']
else:
    ibm_prices = ibm_data['Close']
    print("Warning: 'Adj Close' not found. Using 'Close' prices instead.")

# --- 2. Calculate Log Returns ---
# Log returns are additive and often preferred for financial modeling.
# We drop the first NaN value that results from the diff/shift operation.
log_returns = np.log(ibm_prices / ibm_prices.shift(1)).dropna()
log_returns.name = "log_returns" # Give the series a name

# Calculate absolute log returns for plotting as 'residuals'
absolute_returns = log_returns.abs()
absolute_returns.name = "Absolute Log Returns"

# --- 3. Compute Squared Returns ---
squared_returns = log_returns**2
squared_returns.name = "squared_returns"

# --- 4. Define Window Sizes (k) to Observe Effects ---
window_sizes = [20, 60] # 20-day (approx 1 month) and 60-day (approx 3 months)

# DataFrame to store all volatility estimates for plotting
volatility_df = pd.DataFrame(index=log_returns.index)
volatility_df = pd.concat([volatility_df, absolute_returns], axis=1)

# --- 5. Calculate Left-aligned Rolling Volatility ---
print("\n--- Calculating Left-aligned Volatility ---")
for k in window_sizes:
    # Rolling mean of squared returns (variance)
    # .rolling(window=k) creates a rolling window object
    # .mean() calculates the mean within each window
    # .sqrt() takes the square root to get volatility (standard deviation)
    # .shift(1) aligns the forecast: volatility at time t is based on data up to t-1
    left_aligned_vol = squared_returns.rolling(window=k, min_periods=1).mean().apply(np.sqrt).shift(1)
    col_name = f"Left-aligned MA({k}) Volatility"
    left_aligned_vol.name = col_name
    volatility_df = pd.concat([volatility_df, left_aligned_vol], axis=1)
    print(f"Calculated left-aligned volatility for k={k}")

# --- 6. Calculate Centered Rolling Volatility ---
print("\n--- Calculating Centered Volatility ---")
for k in window_sizes:
    # .rolling(window=k, center=True) makes the window centered around the current point
    # .shift(1) is still applied to align it for plotting as an 'envelope'
    centered_vol = squared_returns.rolling(window=k, center=True, min_periods=1).mean().apply(np.sqrt).shift(1)
    col_name = f"Centered MA({k}) Volatility"
    centered_vol.name = col_name
    volatility_df = pd.concat([volatility_df, centered_vol], axis=1)
    print(f"Calculated centered volatility for k={k}")

# Drop rows with NaN values that result from rolling calculations at the start/end
volatility_df = volatility_df.dropna()

# --- 7. Plotting the Results ---
plt.figure(figsize=(15, 8))

# Define colors and linewidths for the plot
colors = ["darkgray", "darkblue", "darkred", "green", "purple"]
linewidths = [0.5, 1.2, 1.2, 1.2, 1.2] # Absolute returns thin, volatility lines thicker

# Plot each series
for i, column in enumerate(volatility_df.columns):
    plt.plot(volatility_df.index, volatility_df[column],
             label=column,
             color=colors[i % len(colors)], # Cycle through colors
             linewidth=linewidths[i % len(linewidths)]) # Cycle through linewidths

plt.title("IBM Log Returns and Rolling Volatility Envelopes", fontsize=18)
plt.xlabel(None)
plt.ylabel(None)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title=None, loc='upper right', fontsize=10) # Remove legend title
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\n--- First 5 rows of Volatility DataFrame ---")
print(volatility_df.head())
```

```python
#AIG Returns Analysis (Q-Q Plot Fixed)
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, probplot, skew, kurtosis

sns.set_style("whitegrid")

# --- 1. Get Data for AIG ---
aig_data = yf.download("AIG", start="2007-01-01", end="2022-12-31")
aig_prices = aig_data['Close'].dropna()   # Use Adjusted Close

# --- 2. Calculate Log-Returns ---
# Daily
aig_daily_returns = np.log(aig_prices / aig_prices.shift(1)).dropna()

# Monthly
monthly_prices = aig_prices.resample('M').last()
aig_monthly_returns = np.log(monthly_prices / monthly_prices.shift(1)).dropna()

# Quarterly
quarterly_prices = aig_prices.resample('Q').last()
aig_quarterly_returns = np.log(quarterly_prices / quarterly_prices.shift(1)).dropna()

returns_dict = {
    'Daily Returns': aig_daily_returns,
    'Monthly Returns': aig_monthly_returns,
    'Quarterly Returns': aig_quarterly_returns
}

# --- 3. Histograms with Gaussian Fit ---
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('AIG Log-Returns Histograms with Gaussian Fit', fontsize=18)

for ax, (title, returns) in zip(axes, returns_dict.items()):
    sns.histplot(returns, bins=50, kde=False, stat='density',
                 ax=ax, color='darkblue', alpha=0.3)

    mu, sigma = returns.mean(), returns.std()
    x = np.linspace(returns.min(), returns.max(), 100)
    ax.plot(x, norm.pdf(x, mu, sigma), color='navy', linewidth=2, label='Gaussian Fit')

    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Log Return')
    ax.set_ylabel('Density')
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- 4. Q-Q Plots ---
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('AIG Log-Returns Q-Q Plots', fontsize=18)

for ax, (title, returns) in zip(axes, returns_dict.items()):
    probplot(returns.to_numpy().flatten(), dist="norm", plot=ax)
    ax.set_title(title, fontsize=14)
    ax.get_lines()[0].set_color('darkblue')  # Dots
    ax.get_lines()[0].set_alpha(0.5)
    ax.get_lines()[1].set_color('#262626')   # FIX: dark gray line
    ax.get_lines()[1].set_linewidth(1.5)
    ax.set_xlabel('Theoretical Gaussian Quantiles')
    ax.set_ylabel('Sample Quantiles')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


# --- 5. Skewness & Kurtosis vs. Frequency ---
periods = range(1, 61)
aig_skewness, aig_kurtosis = [], []

for l in periods:
    resampled_prices = aig_prices.iloc[::l]  # FIX: use trading days, not calendar
    resampled_returns = np.log(resampled_prices / resampled_prices.shift(1)).dropna()

    aig_skewness.append(float(skew(resampled_returns)))
    aig_kurtosis.append(float(kurtosis(resampled_returns, fisher=True)))  # excess kurtosis

# Skewness plot
plt.figure(figsize=(12, 6))
plt.plot(periods, aig_skewness, color='darkblue', linewidth=2.0)
plt.title('Skewness of AIG Log-Returns vs. Period', fontsize=16)
plt.xlabel('Period [trading days]')
plt.ylabel('Skewness')
plt.xticks(np.arange(0, 61, 10))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Kurtosis plot
plt.figure(figsize=(12, 6))
plt.plot(periods, aig_kurtosis, color='darkred', linewidth=2.0)
plt.title('Excess Kurtosis of AIG Log-Returns vs. Period', fontsize=16)
plt.xlabel('Period [trading days]')
plt.ylabel('Excess Kurtosis')
plt.xticks(np.arange(0, 61, 10))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#Return autocorrelation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# --- 1. Get Data for AIG ---
aig_data = yf.download("AIG", start="2007-01-01", end="2022-12-31", auto_adjust=True)
aig_prices = aig_data['Close']

# --- 2. Calculate Log-Returns and Squared Returns ---
aig_logreturns = np.log(aig_prices / aig_prices.shift(1)).dropna()
aig_squared_returns = aig_logreturns**2

# --- 3. ACF and PACF Plotting ---
def plot_acf_pacf(series, title_suffix="", color="darkblue", lags=30):
    n = len(series)
    mean = series.mean()
    variance = series.var()

    # ACF
    acf_vals = [1.0]
    for lag in range(1, lags + 1):
        cov = np.sum((series[lag:] - mean) * (series[:-lag] - mean)) / n
        acf_vals.append(float(cov / variance))

    # PACF using Yule-Walker
    pacf_vals = [1.0]
    for k in range(1, lags + 1):
        R = np.zeros((k, k))
        for i in range(k):
            for j in range(k):
                R[i, j] = acf_vals[abs(i - j)]
        r = np.array(acf_vals[1:k+1])
        try:
            phi = np.linalg.solve(R, r)
            pacf_vals.append(float(phi[-1]))
        except np.linalg.LinAlgError:
            pacf_vals.append(0.0)  # fallback value

    # Confidence interval
    conf_interval = 1.96 / np.sqrt(n)

    fig, axes = plt.subplots(2, 1, figsize=(12, 10))
    fig.suptitle(f'ACF and PACF of AIG {title_suffix}', fontsize=18)

    axes[0].bar(range(lags + 1), acf_vals, color=color)
    axes[0].axhline(conf_interval, color='gray', linestyle='--')
    axes[0].axhline(-conf_interval, color='gray', linestyle='--')
    axes[0].set_title("ACF")
    axes[0].set_xlabel("Lag")
    axes[0].set_ylabel("ACF")

    axes[1].bar(range(lags + 1), pacf_vals, color=color)
    axes[1].axhline(conf_interval, color='gray', linestyle='--')
    axes[1].axhline(-conf_interval, color='gray', linestyle='--')
    axes[1].set_title("PACF")
    axes[1].set_xlabel("Lag")
    axes[1].set_ylabel("PACF")

    plt.tight_layout()
    plt.show()


# --- 4. Plot Results ---
plot_acf_pacf(aig_logreturns, title_suffix="Log-Returns", color="darkblue")
plot_acf_pacf(aig_squared_returns, title_suffix="Squared Returns", color="darkred")
```
